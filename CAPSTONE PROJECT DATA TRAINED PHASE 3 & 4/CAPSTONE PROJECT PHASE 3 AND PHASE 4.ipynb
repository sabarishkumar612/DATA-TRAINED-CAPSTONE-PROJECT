# PHASE -3 - EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_csv('IMDB1.csv')
df1
```

```python
df2=pd.read_csv('IMDB2.csv')
df2
```

```python
#COMBINING TWO DATAFRAMES USING JOIN OPERATION
ds1=df1.join(df2, lsuffix='_left', rsuffix='_right')
ds1
```

```python
#REMOVING THE UNNECESSARY COLUMNS
ds1.drop(['MOVIE_NAME_right','Unnamed: 0_left','Unnamed: 0_right'], axis=1,inplace=True)
ds1
```

```python
#REPLACING HYPHENS '-' WITH NAN
ds1[ds1.columns.drop(['S.NO','MOVIE_NAME_left','DIRECTORS','DURATION IN MINUTES','YEAR','RATING','STARS','VOTES','GENRE'])]=ds1[ds1.columns.drop(['S.NO','MOVIE_NAME_left','DIRECTORS','DURATION IN MINUTES','YEAR','RATING','STARS','VOTES','GENRE'])].replace('-', np.nan)
ds1
```

```python
#CHECKING FOR NULL VALUES IN EACH COLUMNS
ds1.isnull().sum()
```

```python
#DROPING THE ROWS WITH NULL VALUES IN GROSS COLLECTION COLUMN AS IT IS A TARGERT VARIABLE
ds1 = ds1.dropna(subset=['GROSS COLLECTION'])
ds1
```

```python
#FINDING MODE OF CERTIFICATION COLUMN
print(ds1['CERTIFICATION'].mode())
```

```python
#REPLACING THE NULL VALUES IN CERTIFICATION COLUMN WITH IT'S MODE
ds1['CERTIFICATION'].fillna(ds1['CERTIFICATION'].mode()[0],inplace=True)
```

```python
#CHANGING THE DATA TYPE OF NUMERICAL COLUMNS
ds1['GROSS COLLECTION'] = pd.to_numeric(ds1['GROSS COLLECTION'],errors='coerce')
ds1['POPULARITY'] = pd.to_numeric(ds1['POPULARITY'],errors='coerce')
ds1['META_SCORE'] = pd.to_numeric(ds1['META_SCORE'],errors='coerce')
```

```python
#CHECKING FOR THE DATA TYPES
ds1.dtypes
```

```python
#REPLACING THE NULL VALUES IN NUMERICAL COLUMNS WITH IT'S MEAN
ds1['META_SCORE'].fillna(ds1['META_SCORE'].mean(),inplace=True)
ds1['POPULARITY'].fillna(ds1['POPULARITY'].mean(),inplace=True)
```

```python
#CHECKING FOR NULL VALUS AGAIN
ds1.isnull().sum()
```

NOW IT IS CLEAR THAT THE DATA SET IS FREE FROM NULL VALUES

```python
maxacc=0
maxrs=0
for i in range(1,200):
    x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.3,random_state=i)
    LR=GradientBoostingRegressor()
    LR.fit(x1_train,y1_train)
    pred=LR.predict(x1_test)
    acc=r2_score(y1_test,pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print(maxacc)
print(maxrs)    
```
